```r
#extract datasets

distr1 <- read.csv(file="../../log/degree_distr2018_08_03_15_29_49.csv", skip=12, header=TRUE, sep=",") ## read file
distr2 <- read.csv(file="../../log/degree_distr2018_08_03_15_30_44.csv", skip=12, header=TRUE, sep=",") ## read file
distr3 <- read.csv(file="../../log/degree_distr2018_08_03_15_31_04.csv", skip=12, header=TRUE, sep=",") ## read file
distr4 <- read.csv(file="../../log/degree_distr2018_08_03_15_31_22.csv", skip=12, header=TRUE, sep=",") ## read file
distr5 <- read.csv(file="../../log/degree_distr2018_08_03_15_31_53.csv", skip=12, header=TRUE, sep=",") ## read file
distr6 <- read.csv(file="../../log/degree_distr2018_08_03_15_32_09.csv", skip=12, header=TRUE, sep=",") ## read file
```

```r
degrees <- distr1[,c("degree")]
eps<-0.01
X <- 1:(length(hist(degrees, breaks=20)$breaks) - 1)
Y <- hist(degrees, breaks=20)$counts + eps
X <- log(X)
Y <- log(Y + eps)
DATA = data.frame(X,Y)
linearMod <- lm(Y ~ X, data=DATA)  # build linear regression model on full data
print(linearMod)
print(summary(linearMod))
plot(DATA)
abline(linearMod)
k1<-linearMod$coefficients[2]
```
